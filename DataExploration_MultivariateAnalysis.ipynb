(1) Does the performance in the exams vary according to gender, origin and education level of parents? (Sınavlardaki performans, ebeveynlerin cinsiyetine, kökenine ve eğitim düzeyine göre farklılık gösteriyor mu?)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from scipy.stats.mstats import winsorize
import warnings
df.dropna(inplace=True)
```

```python
a = df["gender"].unique()
```

```python
grup_a =df.groupby("gender")
```

```python
pd.options.display.float_format = '{:.15f}'.format
for var in ["math score", "reading score", "writing score"]:
    comparison = pd.DataFrame(columns=['group_1', 'group_2','statistic', 'p_value'])
    print("For the sake of comparison {}".format(var),end='')
    for i in range(0, len(a)):
        for j in range(i+1, len(a)):
            ttest = stats.ttest_ind(df[df["gender"]==a[i]][var], 
                                df[df["gender"]==a[j]][var])
            group_1 = a[i]
            group_2 = a[j]
            statistic = ttest[0]
            p_value = ttest[1]
            
            comparison = comparison.append({"group_1" : group_1 ,
                                                  "group_2" : group_2 ,
                                                  "statistic": statistic , 
                                                  "p_value" : p_value}, ignore_index=True)
    display(comparison)
```

Aralarında bir ilişki yoktur.
